```python
#Import Dependencies
import pandas as pd

purchase_data = "Resources/heroesofpymoli_purchase_data.csv"

purchase_data = pd.read_csv(purchase_data)
purchase_data.head()
```

```python
#Player Count
players = len(purchase_data["SN"].value_counts())
player_count=pd.DataFrame([players], columns = ["Total Players"])
player_count.head()
```

```python
#Number of Unique Items

UniqueItems = len(purchase_data["Item Name"].unique())

#Average Price
AveragePrice = purchase_data["Price"].mean()

#Number of Purchases
NumberPurchases = purchase_data["Item Name"].count()

#Total Revenue
TotalRevenue = purchase_data["Price"].sum()

#Create Data Frame
Purchasing_Analysis = pd.DataFrame({"Number of Unique Items": [UniqueItems],
                                           "Average Price": [AveragePrice],
                                           "Number of Purchases": [NumberPurchases],
                                           "Total Revenue": [TotalRevenue]})
Purchasing_Analysis = Purchasing_Analysis[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

#Financial Formatting
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${0:,.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${0:,.2f}".format)


Purchasing_Analysis
```

```python
#Player Gender Demographics
Player_Gender = purchase_data["Gender"].value_counts()
Player_Gender_df = pd.DataFrame(Player_Gender)

# Calculate Percentage of Players by Gender
percent_of_players = (purchase_data["Gender"].value_counts() / (purchase_data["Gender"].count())) * 100

# Add 'Percentage of Players' to table
Player_Gender_df["Percentage of Players"] = percent_of_players
gender_demographics_df = pd.DataFrame(Player_Gender_df)

# Rename 'Gender' to 'Total Count'
updated_gender_demographics_df = gender_demographics_df.rename(columns={
    "Gender": "Total Count",
})
#Add Percent to output
updated_gender_demographics_df["Percentage of Players"] = updated_gender_demographics_df["Percentage of Players"].map("{:.2f}%".format)

updated_gender_demographics_df
```

```python
#Gender Purchasing Analysis
#Calculations
gender_purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchasing Price")
gender_counts = purchase_data.groupby(["Gender"]).count()["Price"].rename("Total Gender")
Gender_average= purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Price")

#Create data frame
gender_demographics_df = pd.DataFrame({"Purchase Count": gender_counts, 
                                    "Average Purchase Price": Gender_average,
                                    "Average Purchase Total per Person": gender_purchase_total})


gender_demographics_df
```

```python
#Age Demographics
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Assign bin groups based on age
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=groups)
total_age_count = purchase_data["Age"].value_counts()

#Add percentages
age_percentage = (total_age_count / players) * 100
#Present information in data frame
age_demographics = pd.DataFrame({"Percentage of Players": age_percentage, "Total Count": total_age_count})
age_demographics
 
```

```python
#Age purchase analysis
#Perform Calculations
age_purchase_total = purchase_data.groupby(["Age"]).sum()["Price"].rename("Total Purchasing Price")
age_counts = purchase_data.groupby(["Age"]).count()["Price"].rename("Total Age")
age_purchase_average= purchase_data.groupby(["Age"]).mean()["Price"].rename("Average Price")
#Create data frame
age_demographics_df = pd.DataFrame({"Purchase Count": age_counts, 
                                    "Average Purchase Price": age_purchase_average,
                                    "Average Purchase Total per Person": age_purchase_total})


age_demographics_df
```

```python
#Top Spenders
#Calculations
top_purchase_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchasing Price")
top_purchase_counts = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
top_purchase_average= purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Price")
#Create data frame
top_spenders_df = pd.DataFrame({"Purchase Count": top_purchase_counts, 
                                    "Average Purchase Price": top_purchase_average,
                                    "Average Purchase Total per Person": top_purchase_counts})


top_spenders_df.head()
```

```python
#Most Popular Items
#Extract information
item_info = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
#Perform Calculations
item_purchase_total = item_info.groupby(["Item Name"]).sum()["Price"].rename("Total Purchase Value")
item_purchase_average = item_info.groupby(["Item Name"]).mean()["Price"].rename("Item Price")
item_count = item_info.groupby(["Item Name"]).count()["Price"].rename("Purchase Count")
#Create data frame
item_info_df = pd.DataFrame({"Total Purchase Value": item_purchase_total, "Item Price": item_purchase_average, "Purchase Count": item_count})
item_info_df
item_info_df.head()
```

```python
#Most Profitable Items
#Extract information
item_info = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
#Perform Calculations
item_purchase_total = item_info.groupby(["Item Name"]).sum()["Price"].rename("Total Purchase Value")
item_purchase_average = item_info.groupby(["Item Name"]).mean()["Price"].rename("Item Price")
item_count = item_info.groupby(["Item Name"]).count()["Price"].rename("Purchase Count")
#Insert Calculations into data frame
item_info_df = pd.DataFrame({"Total Purchase Value": item_purchase_total, "Item Price": item_purchase_average, "Purchase Count": item_count})
item_info_df.sort_values(by="Total Purchase Value", ascending=False)

item_info_df
item_info_df.head()
```
